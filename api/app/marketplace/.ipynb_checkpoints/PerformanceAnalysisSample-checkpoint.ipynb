### Wealth index & drawdown chart from Yahoo daily data - Sample analysis
Plus key summary stats for each instrument and optionally for the portfolio. If equity weights are provided then portfolio performance is calculated either. 


    Arguments of the perfAnalysis function:
        eq_tickers - the list of tickers to analyze
        start - start date of the analysis period
        end - end date of the analysis period
        riskfree_rate - riskfree rate approximation for the Sharpe ratio calculation
        portf_weights - the weights of the analyzed equities. If no weigths are provided the portfolio won't be calculated
                        The number of tickers must match to the number of portfolio weights 
                        The weights must add up to 1
        init_cap - initial capital for the wealth index calculation
        chart_size
